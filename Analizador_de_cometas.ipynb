```python
# Librerías
import pandas as pd
import plotly.graph_objects as go

# Módulos
from Funciones.Buscar_cometa import buscar_cometa
from Funciones.Envolvente_superior_inferior import envolvente_superior_inferior
from Funciones.Envolvente_superior_inferior_archivos import envolvente_superior_inferior_archivos
from Funciones.Generar_grafica import generar_grafica
```

```python
# Panel de control
nombre_del_cometa = '12P'
fecha_inicial = '1990-06-01'
```

```python
buscar_cometa(nombre_del_cometa)
```

```python
# Base de datos
bases_de_datos_cobs = envolvente_superior_inferior(nombre_del_cometa, fecha_inicial)
```

```python
tipo_de_grafica = '1' 
generar_grafica(tipo_de_grafica, nombre_del_cometa, bases_de_datos_cobs)
```

```python
# Cargar archivo con los datos del MPC
datos_mpc_df = pd.read_parquet('Archivos/12P_Procesado_MPC.parquet')
datos_mpc_df
```

```python
bases_de_datos_mpc =  envolvente_superior_inferior_archivos(nombre_del_cometa, datos_mpc_df)
```

```python
tipo_de_grafica = '7' 
generar_grafica(tipo_de_grafica, nombre_del_cometa, bases_de_datos_mpc)
```

```python
curva_cruda_cobs_df = bases_de_datos_cobs[1]
curva_cruda_mpc_df = bases_de_datos_mpc[1] 

envolvente_cobs_df = bases_de_datos_cobs[2] 
nucleo_cobs_df = bases_de_datos_cobs[3]
envolvente_mpc_df = bases_de_datos_mpc[2] 
nucleo_mpc_df = bases_de_datos_mpc[3]
```

```python
fig = go.Figure()
    
fig.add_trace(go.Scatter(x=curva_cruda_cobs_df.obs_date, 
        y=curva_cruda_cobs_df.magnitud_reducida, mode='markers',
        name='Cruda_Cobs', 
        marker=dict(color='#fa04f6', line=dict(width=1, color='DarkSlateGrey'))))

fig.add_trace(go.Scatter(x=envolvente_cobs_df.obs_date, 
        y=envolvente_cobs_df.promedio_movil, mode='markers',
        name='Envolvente_Cobs', 
        marker=dict(color='yellow', line=dict(width=1, color='DarkSlateGrey'))))

fig.add_trace(go.Scatter(x=nucleo_cobs_df.obs_date, 
        y=nucleo_cobs_df.promedio_movil, mode='markers',
        name='Nucleo_Cobs', 
        marker=dict(color='red', line=dict(width=1, color='DarkSlateGrey'))))

fig.add_trace(go.Scatter(x=curva_cruda_mpc_df.obs_date, 
        y=curva_cruda_mpc_df.magnitud_reducida, mode='markers',
        name='Cruda_MPC', 
        marker=dict(color='#04ebfa', line=dict(width=1, color='DarkSlateGrey'))))

fig.add_trace(go.Scatter(x=envolvente_mpc_df.obs_date, 
        y=envolvente_mpc_df.promedio_movil, mode='markers',
        name='Envolvente_MPC', 
        marker=dict(color='#fa7004 ', line=dict(width=1, color='DarkSlateGrey'))))

fig.add_trace(go.Scatter(x=nucleo_mpc_df.obs_date, 
        y=nucleo_mpc_df.promedio_movil, mode='markers',
        name='Nucleo_MPC', 
        marker=dict(color='#fa045b', line=dict(width=1, color='DarkSlateGrey'))))

fig.update_layout(template='plotly_dark')
fig.update_yaxes(autorange="reversed")
fig.update_layout(template='plotly_dark', xaxis_title='Observation Date', yaxis_title='Averaged Magnitude', title = f'Max/Min Averaged Lightcurve of comet {nombre_del_cometa}')
fig.show()
```
